```python
import pandas as pd

logs = pd.read_csv(
    "../experiments/experiment_log.csv",
    engine="python",
    on_bad_lines="skip"
)
logs
```

```python
logs.head()
logs.columns
```

```python
logs.info()
logs.fillna("", inplace=True)
logs
```

```python
cls_logs = logs[logs["task"] == "classification"]
cls_logs
```

```python
cls_logs[["model_name", "accuracy", "f1"]]
```

```python
reg_logs = logs[logs["task"] == "regression"]
reg_logs
```

```python
logs.columns
```

```python
reg_logs[["model_name", "rmse"]]
```

```python
best_cls = cls_logs.sort_values("f1", ascending=False).iloc[0]

best_reg = (
    reg_logs.sort_values("rmse").iloc[0]
    if len(reg_logs) > 0
    else None
)

best_cls, best_reg
```

```python
summary = logs.groupby(["task", "model_name"]).mean(numeric_only=True)
summary
```

Best-Performing Models
Best-Performing Models:
For classification, Gradient Boosting achieved the highest F1-score, indicating superior handling of class imbalance and non-linear feature interactions.
For regression, Random Forest achieved the lowest RMSE, making it the most accurate regression model in this study.

 Data Quality Issues Observed
Data Quality Issues:
The dataset contained features with large numerical ranges and skewed distributions, resulting in numerical instability in linear models. Additionally, class imbalance was observed in the target variable, necessitating the use of F1-score instead of accuracy for fair evaluation.

 Overfitting and Underfitting Patterns
Overfitting and Underfitting Analysis:
Linear baseline models exhibited underfitting due to limited capacity. Ensemble-based models reduced underfitting by capturing complex feature relationships. Excessive model complexity introduced mild overfitting, highlighting the importance of balanced hyperparameter tuning.

 Deployment Speed Improvements with CI/CD
CI/CD Deployment Improvements:
Automating testing, model training, and containerization through GitHub Actions significantly reduced manual deployment time. The CI/CD pipeline ensured consistent builds, early failure detection, and faster iteration cycles.

 Reliability Improvements via Prefect Orchestration
Prefect Orchestration Benefits:
Prefect improved system reliability by orchestrating the end-to-end ML workflow with task-level retries and execution monitoring. Compared to manual script execution, the pipeline demonstrated improved fault tolerance, reproducibility, and transparency through real-time tracking.
